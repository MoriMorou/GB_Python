```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics


%matplotlib inline
pd.set_option('display.max_rows', 40)
```

```python
# Импорт данных
df = pd.read_csv('time_line.csv', sep=';')
df.head()
```

```python
df.info()
```

```python
# Подготовка данных
df = df[['Значение', 'Дата']]
df['Дата'] = pd.DatetimeIndex(df['Дата'],dayfirst=True)
df['Месяц'] = pd.DatetimeIndex(df['Дата']).month 
df['Год'] = pd.DatetimeIndex(df['Дата']).year
data = df.groupby(['Год', 'Месяц'], as_index=False).sum()
data['index'] = data.index
```

```python
df.head()
```

```python
# Строим гистограммы по каждому признаку: 
sns.pairplot(data[['Значение','index', 'Год', 'Месяц']], hue = 'Год')
```

```python
# Хорошо заметен линейний низсходящий тренд.
# Теперь посмотрим на математические значения зависимостей:
data[['Значение','index', 'Год', 'Месяц']].corr()
```

Значения коэффициента корреляции интерпретируются следующим образом:

- До 0,2 — очень слабая корреляция
- До 0,5 — слабая
- До 0,7 — средняя
- До 0,9 — высокая
- Больше 0,9 — очень высокая

Действительно видим, что между переменными «index» и «Год» выявлена очень сильная зависимость 0.97.
