```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import *
from datetime import datetime, timedelta, date
from dateutil import relativedelta as rd
```

```python
# Upload DB
orders = pd.read_csv("orders_20190822.csv",sep=";",decimal=",",parse_dates=["o_date"])
```

```python
list(orders.columns)
```

```python
# Rename columns
orders['dateOr'] = pd.to_datetime(orders.date)
orders = orders.rename(columns={'id_o': 'id', 'user_id': 'user'})
list(orders.columns)
```

```python
# We translate into date format and delete fake orders
orders = orders[~(orders['price']<=99) & ~(orders['price']>=300000)]
```

```python
price_date = orders[["date","price"]]
price_date.head()
```

```python
clients = orders.groupby(['id'], as_index=False)['date'].idxmin()
clients.head()
```
