#### 1. Задание
Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20). Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
```

```python
a = np.array([20, 20, 20])
b = np.array([0, 0, -20])
```

```python
def vector_cal(vector):
    length_vec = np.sqrt(np.sum(np.square(vector)))
    return length_vec
```

```python
print(f'Length of vector {a}: {(vector_cal(a))}')
print(f'Length of vector {b}: {(vector_cal(b))}')
```

#### 2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

Ответ: 
В аналитическом выражении прямые, заданные линейными функциями y=ax+b и y=kx+m будут перпендикулярны, если выполнено следующее условие на их угловые коэффициенты ak=-1
То есть 3*(-1/3)=-1, прямые перпендикулярны. Но так как пределы y = от -15 до 15, a x = -4 до 4, то они выглядят не перендикулярными.

```python
# задаем новые пределы
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlim([-15, 15])
plt.ylim([-15, 15])
plt.xlabel('x')
plt.ylabel('y')
```

#### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
- окружности,
- эллипса,
- гиперболы.


```python
#  A - center and R - radius
def circle(A = [0, 0], R = 5):
    a, b = A
    x = np.linspace((a - R), (a + R), 121)
    y1 = np.sqrt(np.power(R, 2) - np.power((x - a), 2)) + b
    y2 = -1 * y1
    plt.plot(x, y1, color='blue')
    plt.plot(x, y2, color='blue')
    plt.xlim([(a - R - 1), (a + R + 1)])
    plt.ylim([(b - R - 1), (b + R + 1)])
    plt.show()
```

```python
# Почему такой кривой круг? (не хнаю как красиов нарисовать)
circle()
```

```python
#  A - center and a, b - major and minor axis
def ellipse(A = [0, 0], a = 10, b = 5):
    x0, y0 = A
    x = np.linspace((x0 - a), (x0 + a), 121)
    y1 = np.sqrt(np.power(b, 2) - np.power(b, 2) *  np.power((x - x0), 2) / np.power(a, 2)) + y0
    y2 = -1 * y1
    plt.plot(x, y1, color='blue')
    plt.plot(x, y2, color='blue')
    plt.xlim([(x0 - a - 1), (x0 + a + 1)])
    plt.ylim([(y0 - b - 1), (y0 + b + 1)])
    plt.show()
```

```python
ellipse()
```

```python
def hyperbola(x = np.linspace(-5, 5, 121), a = 1, b = 1):
    y1 = np.sqrt((x * b / a - b) * ((x * b / a + b)))
    y2 = -1 * y1
    plt.plot(x, y1, color='blue')
    plt.plot(x, y2, color='blue')
    plt.show()
```

```python
hyperbola()
```
