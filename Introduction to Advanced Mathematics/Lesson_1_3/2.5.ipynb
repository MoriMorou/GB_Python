```python
%matplotlib inline


import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
print(X)
X, Y = np.meshgrid(X, Y)
print(Y)
print(X[1,1])
Z = np.sqrt(X**2 + Y**2)
Z2 = 2*X + 3*Y
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-5, 5, 2)

Z = 2*X + 3*Y
ax.plot(X, Y, Z, 'red')
ax.plot(X, Y, 2*Z)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```
