```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 3, 201)
print(np.poly([1, 1, 2]))
plt.plot(x, x**3 - 4*x**2 + 5*x - 3)
plt.plot(x,np.polyval([1., -4.,  5., -2.],x))
plt.plot(x, np.exp(x))
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-1,3) 
plt.ylim(-2,2)
plt.grid(True)
plt.show()
print(np.roots([ 1., -4.,  5., -2.]))
```
