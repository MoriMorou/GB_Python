```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.optimize import fsolve

def f(x):
    return (np.sin(x + 1))

x0 =  fsolve(f, 1)
print (x0)

x = np.linspace(-3*np.pi, 3*np.pi, 201)
plt.plot(x, f(x))
plt.plot([-10, 10],[0,0], color ='red', linewidth=0.75, linestyle ="--") 
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) - x*y - 1)

x1, y1 =  fsolve(equations, (-2, 1))
              
print (x1, y1)
```
