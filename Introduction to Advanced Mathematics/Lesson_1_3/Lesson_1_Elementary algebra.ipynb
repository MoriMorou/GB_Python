#### 4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
x = np.linspace(0, 20, 60)
```

```python
def y(k, x):
    return np.cos(k*x)
```

```python
plt.plot(x, y(1, x), label='k=1', color='green')
plt.plot(x, y(2, x), label='k=2', color='blue')
plt.xlabel('x')
plt.ylabel('y(k,x)=cos(k*x)')
plt.legend()
```
