## Матрицы и векторы
### Основные операции с векторами
### $
(a_{0},\ldots, a_{n})\qquad
(b_{0},\ldots, b_{n})\qquad
$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1, 2, 3, 4, 5], float)
b = np.array([0, 9, 8, 7, 6], float)
a * 2
```

```python
np.outer(a, b)
```

```python
a = np.array([3, 4], float)
U, V = a[0], a[1]

plt.quiver(U, V, angles='xy', scale_units='xy', scale=1)

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid(True)
plt.show()
```

```python
print(a)
np.dot(a, a)
```

```python
np.linalg.norm(a)
```

```python
X, Y = np.array([0, 0, 0]), np.array([0, 0, 0])
U, V = np.array([5, 3, 3]), np.array([4, 2, 0])

plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.show()
```

```python
X, Y = np.array([0, 2, 0]), np.array([0, 3, 0])
U, V = np.array([2, 3, 5]), np.array([3, 1, 4])

plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.show()
```

```python
a = np.array([2, 3], float)
b = np.array([4, 1], float)
X, Y = np.array([0, 0]), np.array([0, 0])
U, V = np.array([a[0], b[0]]), np.array([a[1], b[1]])

plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.show()
```

#### Скалярное произведение
### $(\vec a \cdot \vec b)  = a \cdot b  \cdot \cos \alpha $

```python
s = np.inner(a, b)
s/(np.linalg.norm(a)*np.linalg.norm(b))
```

```python
i = np.array([1, 0], float)
j = np.array([0, 1], float)
X, Y = np.array([0, 0]), np.array([0, 0])
U, V = np.array([i[0], j[0]]), np.array([i[1], j[1]])

plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.show()
np.inner(i, j)
```

#### Векторное произведение
### $\vec a \times \vec b  = a \cdot b  \cdot \sin \alpha \cdot \vec k$

```python
print(a)
print(b)
v = np.cross(a, b)
print(v)
v/(np.linalg.norm(a)*np.linalg.norm(b))
```

#### Векторное произведение
### $\vec a \times \vec b  = \left|
\begin{array}{cccc}
a_{x} & a_{y}\\
b_{x} & b_{y}
\end{array}
\right| \cdot \vec k
$
### $\vec a \times \vec b  = \left|
\begin{array}{cccc}
\vec i & \vec j & \vec k \\
a_{x} & a_{y} & a_{z}\\
b_{x} & b_{y} & b_{z}
\end{array}
\right|
$

#### Смешанное произведение
### $( \vec a \times \vec b ) \cdot c = a \cdot ( \vec b \times \vec c ) $

```python
a = np.array([2, 3, 0], float)
b = np.array([4, 1, 0], float)
c = np.array([2, 6, 1], float)
v = np.cross(a, b)
print (np.inner(v, c))
w = np.cross(b, c)
print (np.inner(w, a))
```
