## Системы линейных алгебраических уравнений (СЛАУ) 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Линейные системы уравнений с прямоугольной матрицей
## $A = \left(
\begin{array}{cccc}
a_{00} & a_{01} & \ldots & a_{0n}\\
a_{10} & a_{11} & \ldots & a_{1n}\\
\vdots & \vdots & \ddots & \vdots\\
a_{m0} & a_{m1} & \ldots & a_{mn}
\end{array}
\right) ,     X = \left(
\begin{array}{cccc}
x_0\\x_1\\\vdots\\x_{n}
\end{array}
\right)
,     B = \left(
\begin{array}{cccc}
b_0\\b_1\\\vdots\\b_{m}
\end{array}
\right)$
## $A\cdot X = B$
### Пример: переопределенные СЛАУ
## $x + 2y = 4$
## $3x - 4y = 1$
## $8x - 5y = 10$


```python
x = np.linspace(1.6, 2.2, 201)
plt.plot(x, 2 - 0.5 * x)
plt.plot(x, 0.75 * x - 0.25)
plt.plot(x, 1.6 * x - 2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## $\left(
\begin{array}{cccc}
1 & 2\\
3 & -4\\
8 & -5\end{array}
\right)\cdot \left(
\begin{array}{cccc}
x\\y\end{array}
\right) = \left(\begin{array}{cccc}
4\\1\\10\end{array}\right)
$

```python
A = np.array([[1, 2], [3, -4], [8, -5]])
B = np.array([4, 1, 10])
np.linalg.lstsq(A, B)
```

```python
np.dot(A, [1.94578313,  1.13975904]) - B
```

```python
def Q(x, y):
    return ((np.linalg.norm(np.dot(A, [x,  y]) - B))**2)
Q(1.95578313,  1.12975904)
```

## $\left(
\begin{array}{cccc}
1 & 2\\
3 & -4\\
4 & -2\end{array}
\right)\cdot \left(
\begin{array}{cccc}
x\\y\end{array}
\right) = \left(\begin{array}{cccc}
4\\1\\5\end{array}\right)
$

```python
x = np.linspace(1.6, 2.1, 201)
plt.plot(x, 2 - 0.5 * x)
plt.plot(x, 0.75 * x - 0.25)
plt.plot(x, 2 * x - 2.5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
A = np.array([[1, 2], [3, -4], [4, -2]])
B = np.array([4, 1, 5])
np.linalg.lstsq(A, B)
```

### Пример: недоопределенные СЛАУ
## $x + 2y = 4$
## $\left(
\begin{array}{cccc}
1 & 2\\
\end{array}
\right)\cdot \left(
\begin{array}{cccc}
x\\y\end{array}
\right) = \left(\begin{array}{cccc}
4\\\end{array}\right)
$

```python
x = np.linspace(-1, 5, 201)
plt.plot(x, 2 - 0.5 * x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
