## Системы линейных алгебраических уравнений (СЛАУ) 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Недоопределенные, вырожденные  и плохо обусловленные системы
## $A = \left(
\begin{array}{cccc}
a_{00} & a_{01} & \ldots & a_{0n}\\
a_{10} & a_{11} & \ldots & a_{1n}\\
\vdots & \vdots & \ddots & \vdots\\
a_{m0} & a_{m1} & \ldots & a_{mn}
\end{array}
\right) ,     X = \left(
\begin{array}{cccc}
x_0\\x_1\\\vdots\\x_{n}
\end{array}
\right)
,     B = \left(
\begin{array}{cccc}
b_0\\b_1\\\vdots\\b_{m}
\end{array}
\right)$
## $A\cdot X = B$
### Пример 1: недоопределенные СЛАУ
## $x + 2y = 12$

```python
x = np.linspace(-1, 5, 201)
plt.plot(x, 6 - 0.5 * x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

### Пример 2: вырожденные СЛАУ
## $x + 2y = 10$
## $1.1x + 2.2y = 14$

```python
A = np.array([[1, 2], [1.1, 2.2]])
B = np.array([10, 14])
x = np.linspace(-1, 5, 201)
plt.plot(x, 6 - 0.5 * x)
plt.plot(x, 14/2.2 - 0.5 * x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
np.linalg.lstsq(A, B)
```

### Пример 3: плохо обусловленные СЛАУ
## $x + 2y = 3$
## $2x + 4.1y = 5$

```python
A = np.array([[1, 2], [2, 4.1]])
B = np.array([3, 5])
x = np.linspace(-1, 500, 201)
plt.plot(x, 3/2 - 0.5 * x)
plt.plot(x, 5/4.1 - (2/4.1) * x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
np.linalg.solve(A, B)
```

## $x + 2y = 3$
## $2.1x + 4y = 5$

```python
D = np.array([[1, 2], [2.01, 4]])
B = np.array([3, 5])
np.linalg.solve(D, B)
```

## Нормальное псевдорешение
## $A\cdot X = B$
## $| X | = min$
### Пример 1: недоопределенные СЛАУ
## $x + 2y = 12$
## $X = \left(
\begin{array}{cccc}
x\\y\end{array}
\right)$
## $\left(
\begin{array}{cccc}
1 & 2\\
\end{array}
\right)\cdot X = \left(\begin{array}{cccc}
12\\\end{array}\right)
$

```python
def Q(x, y):
    return (x**2 + y**2)

x = np.linspace(-1, 5, 201)
plt.plot(x, Q(x, 6 - 0.5 * x))
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
A = np.array([[1, 2]])
B = np.array([12])
np.linalg.lstsq(A, B)
```
