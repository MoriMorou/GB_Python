#Базовые определения

##Основные определения

**Договор имущественного страхования** (ГК РФ Статья 929):


> По договору имущественного страхования **страховщик** обязуется за обусловленную договором плату (**страховую премию**) при наступлении предусмотренного в договоре события (**страхового случая**) возместить другой стороне (**страхователю**) или иному лицу, в пользу которого заключен договор (**выгодоприобретателю**), причиненные вследствие этого события **убытки** в застрахованном имуществе либо **убытки** в связи с иными имущественными интересами **страхователя** - выплатить **страховое возмещение** в пределах определенной договором **страховой суммы**.


    

**Страховщик** (ГК РФ Статья 938):
> Юридическое лицо, имеющее разрешения (лицензии) на осуществление страхования соответствующего вида.

**Существенные условия договора страхования** (ГК РФ Статья 942):
> При заключении договора **имущественного страхования** между **страхователем** и **страховщиком** должно быть достигнуто соглашение:
* об определенном **имуществе** либо ином **имущественном интересе**, являющемся **объектом страхования**;
* о характере события **страхового случая**, на случай наступления которого осуществляется страхование;
* о размере **страховой суммы**;
* о **сроке действия** договора.

**Страховая сумма** (ГК РФ Статья 947):
> Сумма, в пределах которой **страховщик** обязуется выплатить **страховое возмещение** по договору **имущественного страхования**, определяется соглашением **страхователя** со **страховщиком**.<br/>
При страховании имущества, если договором страхования не предусмотрено иное, "страховая сумма" не должна превышать действительную "страховую стоимость".<br/>
Такой стоимостью считается для имущества его действительная стоимость в месте его нахождения в день заключения договора страхования.

**Премия (по договору)** (ГК РФ Статья 954):
> Сумма, которую страхователь обязан уплатить страховщику по КАСКО включает тариф за ущерб, тариф за тоталь, тариф за угон и комиссию агента.

##Прочие определения

**Подписанная премия (по портфелю)**:
> Сумма премий по договорам, заключенным в отчетном периоде.

**Экспозиция (по договору)** (Актуарные правила №1):
> Выраженный в годах срок действия договора страхования в отчетном периоде.

**Экспозиция (по портфелю)**:
> Сумма экспозиций по договорам, заработанных за отчетный период по всем договорам, действовавшим хотя бы один день в отчетном периоде.

**Заработанная премия (по договору)**:
> Часть **страховой премии**, которая признается доходом страховщика в отчетном периоде и в случае аннулирования или расторжения договора страхования не подлежит возврату страхователю.<br/> Заработанная премия исчисляется от общей **страховой премии** как пропорция по времени.<br/> В течение каждого дня действия годового страхового полиса страховщик зарабатывает 1/365 часть **страховой премии**.

**Заработанная премия (по портфелю)**:
> Сумма премий, заработанных за отчетный период по всем договорам, действовавшим хотя бы один день в отчетном периоде. 

**Частота убытков** (Актуарные правила №1):
> Количество произошедших страховых случаев в расчете на один годовой договор, находится делением числа произошедших страховых случаев на экспозицию по договорам.

**Убыточность (актуарная)**:
> Отношение **страховых выплат** и сумм, направленных на увеличение объема **страховых резервов**, в отчетном периоде к величине **заработанной премии** за отчетный период. 

**Убыточность (андеррайтерская)**:
> Отношение **страховых выплат** и сумм, направленных на увеличение объема **страховых резервов**, по договорам, заключенным в отчетном периоде, к величине **заработанной премии** за отчетный период.

  ![](https://drive.google.com/uc?export=view&id=1rkzVglYfkC92pAKIImvGhhFnkVbWUm9U)

1, 2, 3, 4, 5 - полисы

A, B, C, D, E - убытки 

**Подписанная премия** (по продажам в месяце 1) =  ( Премия_1 + Премия_2 + Премия_3 )

**Подписанная премия** (по продажам в месяце 2) = ( Премия_4 + Премия_5 ) 

**Убыточность** (андеррайтерская, по продажам в месяце 1) = ( A + B + C + D ) / ( Премия_1 + Премия_2 + Премия_3 )

**Убыточность** (андеррайтерская, по продажам в месяце 2) = ( E ) / ( Премия_4 + Премия_5 )


**Экспозиция** (месяц 3) = 1/12 * 5 = 5/12

**Заработанная премия** (месяц 3) = ( Премия_1 * 1/12 + Премия_2 * 1/12 + Премия_3 * 1/12 + Премия_4 * 1/12 + Премия_5 * 1/12 ) 

**Частота** (месяц 3) = 2 / (5/12) = 24/5 = 480%

**Убыточность** (актуарная, по действию в месяце 3) = ( C + E ) / Заработанная премия (месяц 3)

##Проверка качества заполнения данных

Поскольку freMPL состоит из 10 отдельных датасетов, каждый из которых имеет различные переменные, есть смысл заранее посмотреть, какие датасеты стоит использовать, чтобы затем не возникало проблем с большим количеством пропущенных данных.

```
# Смотрим, какие факторы пропущены в каждом из датасетов

dct = {}
for i in range(1,11):
    _x = df.loc[df.Dataset == i].notnull().sum()
    dct[i] = list(_x[_x == 0].index)

print('Dataset  Missing Variables')
for x in range(1,11):
    print(x,'\t',dct[x])
```

Наборы данных 1-4 не содержат информации о количестве страховых требований, а 5-10 не содержат информации по характеристикам транспортного средства. Тем не менее, наборы 5-9 имеют одинаковые пропущенные факторы, имеет смысл попробовать их объединить.

```
# Объединяем наборы данных 5-9, удаляем пустые столбцы, удаляем дубликаты

df59 = df.loc[df.Dataset.isin([5, 6, 7, 8, 9])]
df59 = df59.drop(['Dataset'], axis=1)
df59 = df59.dropna(axis=1, how='all')
print('   With duplicates', len(df59))
df59 = df59.drop_duplicates()
print('Without duplicates', len(df59))
```

Внутри каждого набора данных имеется около 10 000 наблюдений дубликатов. Все дубликаты были удалены.

```
# Вспомогательный столбец для суммирования числа полисов

df['PolicyCount'] = 1
```

```
# Вспомогательный столбец для суммирования числа полисов с убытками

df['ClaimCount'] = df['ClaimAmount'] > 0
```
