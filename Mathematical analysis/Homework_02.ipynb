```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### 1. Предложить пример функции,  не имеющей предела в нуле и в бесконечностях.

```python
def f(x):
    if np.fabs(x) < 1:
        return np.sin(1/x)
    else:
        return np.sin(x)
    
x = np.linspace(-10, 10, 200)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

### 2. Привести пример функции, не имеющей предела в точке, но определенной в ней.

```python
def f(x):
    if x < 0:
        return 0
    else:
        return 1
    
x = np.linspace(-100, 100, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

### 3. Исследовать функцию $f(x) = x^3 - x^2$ по плану:
1. Область задания и область значений
2. Нули функции и их кратность
3. Отрезки знакопостоянства
4. Интервалы монотонности
5. Четность функции
6. Ограниченность
7. Периодичность

```python
# постороим функцию

def f(x):
    return np.power(x, 3) - np.power(x, 2)

x = np.linspace(-100, 100, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

### 4. Найти предел:
$\lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2}$<br>

$\lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2}$ = $\frac{3x^3 - 2x^2}{4x^2}$ = $\frac{x^2(3x - 2)}{4x^2}$ = $\frac{(3x - 2)}{4}$ = $-\frac{1}{2}$

### 5. Найти предел:
$\lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1}$<br>

$\lim_{x \to 0} \frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1}$ = $\frac{\sqrt{1 + x} - 1}{\sqrt[3]{1 + x} - 1}$ = $\frac{x(\sqrt{1 + x} - \frac{1}{x})}{x(\sqrt[3]{1 + x} - \frac{1}{x})}$ = $\frac{\sqrt{1 + x} - \frac{1}{x}}{\sqrt[3]{1 + x} - \frac{1}{x}}$ = $\frac{\frac{1}{2}}{\frac{1}{3}}$ = $\frac{3}{2}$

### 6. Найти предел:
$\lim_{x \to \infty} (\frac{x + 3}{x})^{4x+1}$<br>

Второй змечательный предел:<br>
$\lim_{x \to \infty} (\frac{x + 3}{x})^{4x+1}$ = $(\frac{x + 3}{x})^{4x+1}$ = $(1 + \frac{3}{x})^{4x+1}$ = $e^{12}$

### 10. Найти предел:

$\lim_{x \to \infty} (\frac{4x+3}{4x-3})^{6x}$<br>

Второй змечательный предел:<br>
$\lim_{x \to \infty} (\frac{4x+3}{4x-3})^{6x}$ = $(\frac{4x+3}{4x-3})^{6x}$ = $(1 +\frac{6}{4x-3})^{\frac{4}{4}(6x)}$ = $(1 +\frac{6}{4x-3})^{\frac{6}{4}(4x-3)}$ = $e^{6\frac{6}{4}}$ = $e^9$

### 11. Найти предел:
$\lim_{x \to \infty} \frac{sin(x) + ln(x)}{x}$<br>

$\lim_{x \to \infty} \frac{sin(x) + ln(x)}{x}$ = $\frac{sin(x) + ln(x)}{x}$ = $0 + ln(x)$ = $+\infty$
