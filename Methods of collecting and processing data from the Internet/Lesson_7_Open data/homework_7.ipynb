```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import requests
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
pd.options.display.max_columns = None
import seaborn as sns
%pylab inline
```

```python
link = "http://sbrfdata.ru/opendata.zip"
data = pd.read_csv(link, encoding="cp1251")
data.date = pd.to_datetime(data.date)
data.head()
```

```python
names_list = sorted(list(data['name'].unique()))
regions_list = sorted(list(data['region'].unique()))
```

```python
def date_to_datetime(date):
    return datetime.datetime.fromisoformat(date)

def get_df_by_condition(data, name, regions, later_than_date, before_than_date):
    regions_names = [regions_list[i] for i in regions]
    tmp_df = data.loc[(data.name==names_list[name]) & (data.region.isin(regions_names))].copy()
    return tmp_df.loc[(tmp_df.date > date_to_datetime(later_than_date)) & (tmp_df.date < date_to_datetime(before_than_date))].copy()
```

```python
def show_plot(df):
    pylab.rcParams['figure.figsize'] = (12, 6)
    regions = sorted(list(df['region'].unique()))
    for region in regions:
        tmp_df = df.loc[df.region == region]
        plt.plot(tmp_df.date, tmp_df.value, label=region)
    plt.title(df.name.unique()[0])
    plt.legend()
    plt.show()
```

```python
print("Доступные \"name\"\'s:\n")
for i, name in enumerate(names_list):
    print(i, "-", name)
```

```python
print("Доступные \"region\"\'s:\n")
for i, region in enumerate(regions_list):
    print(i, "-", region)
```

```python
print("Доступные \"date\"\'s:\n")
print("Минимум:", data.date.min())
print("Максимум:", data.date.max())
```

#### Пример вывода данных по региона за определенный период 

```python
tst_df = get_df_by_condition(data=data, name=5, regions=[71, 5, 6, 7], later_than_date="2018-01-15", before_than_date="2019-01-15")
tst_df
```

#### Пример графического вывода данных

```python
show_plot(tst_df)
```
